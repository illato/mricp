## TemplateFlow

---

(_T2-w_ Template For Registration)

---

Among the different data sources, not all have same explicit type of _T1-w_ data.  

---

(__one source__ has _T1-w_ __MPRAGE__, __others__ simply indicate ___T1-w___)  

- To minimize bias
  - Only _T2-w_ imagery was used
    - (rather than registering _T1-w_ data to a _T1-w_ template and subsequently registering each _T2-w_ scan to the corresponding _T1-w_ scan)
      - ___T2-w_ data was directly registered to established an _T2-w_ template__
        - (TemplateFlow)
  

---
